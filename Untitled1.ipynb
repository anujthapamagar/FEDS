```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as sps
```

```python
tem = 'land_global_temperature_anomaly-1850-2023.csv'
year, temp_anomaly = np.loadtxt(tem, delimiter=',', skiprows=5, unpack=True)
```

## Plot the data

```python
plt.plot(year, temp_anomaly);
```

```python
plt.figure(figsize=(10,5))

# Plotting
plt.plot(year, temp_anomaly, color = 'blue', linestyle='-', linewidth=1)
plt.title('Global temperature anomaliies')
plt.xlabel('Year')
plt.ylabel('Land temperature anomaly [Â°C]')
plt.grid();
```

## Linear regression

```python
# Calculate the mean values
year_mean = np.mean(year)
temp_anomaly_mean = np.mean(temp_anomaly)

#a_1 coefficient
a_1 = np.sum(temp_anomaly*(year - year_mean)) / np.sum(year*(year - year_mean))

#a_0 coefficient
a_0 = temp_anomaly_mean - a_1*year_mean

a_1, a_0
```

## lets call `reg` the array obtained from evaluating $f(x)$ for all years

```python
reg = a_0 + a_1 * year
```

```python
plt.figure(figsize = (10,5))
plt.plot(year, temp_anomaly)
plt.plot(year, reg, 'k--', linewidth=2, label='Linear regression')
plt.ylabel('Global temperature anomaly [Â°C]')
plt.legend()
plt.grid();
```

## Apply regression using NumPy

With `polyfit()` we get the slope and $y$-intercept of the line that best fits the data, with `poly1d()` we can build the linear function from its slope and $y$-intercept

```python
# First fit with NumPy, then name the coefficients obtained a_1n, a_0n:
a_1new, a_0new = np.polyfit(year, temp_anomaly, 1)

f_linear = np.poly1d((a_1new, a_0new))
a_1new, a_0new
```

```python
plt.figure(figsize = (10,5))
plt.plot(year, temp_anomaly)
plt.plot(year, f_linear(year),'k--', label='Linear regression')
plt.ylabel('Global temperature anomaly [Â°C]')
plt.legend()
plt.grid();
```

## There seems to be a faster increase in temperature since 1970 so maybe single regression line might not represent whole data.

lets break the data into two before and after 1970.
To do that, we first need to find the position in our year array where the year 1970 is located. Thankfully, NumPy has a function called `numpy.where()` that can help us. We pass a condition and `numpy.where()` tells us where in the array the condition is `True`.

```python
np.where(year==1970)
```

```python
# slicing the data

year1, temp_anomaly_1 = year[0:120], temp_anomaly[0:120]
year2, temp_anomaly_2 = year[120:], temp_anomaly[120:]

m1, b1 = np.polyfit(year1, temp_anomaly_1, 1)
m2, b2 = np.polyfit(year2, temp_anomaly_2, 1)

f_linear_1 = np.poly1d((m1, b1))
f_linear_2 = np.poly1d((m2, b2))
```

```python
plt.figure(figsize=(10,5))
plt.plot(year, temp_anomaly)
plt.plot(year1, f_linear_1(year1), 'g--', label='1850-1970')
plt.plot(year2, f_linear_2(year2), 'r--', label='1970-2023')
plt.ylabel('Global temperature anomaly [Â°C]')
plt.legend()
plt.grid();
```
